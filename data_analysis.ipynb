```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('user_data.csv')

# Convert age and total income to numeric values
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['total_income'] = pd.to_numeric(df['total_income'], errors='coerce')

# Visualizations
# Ages with the highest income
top_income_by_age = df.groupby('age')['total_income'].sum().sort_values(ascending=False).head(10)
top_income_by_age.plot(kind='bar', title='Top 10 Ages with Highest Income')
plt.show()

# Gender distribution across spending categories
expense_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
for category in expense_categories:
    if category in df.columns:
        df[category] = pd.to_numeric(df[category], errors='coerce')
    else:
        df[category] = 0  # or some default value if the column is missing
        
gender_spending = df.groupby('gender')[expense_categories].sum()
gender_spending.plot(kind='bar', title='Gender Distribution Across Spending Categories')
plt.show()

# Exporting the charts
plt.savefig('gender_spending.png')
plt.savefig('top_income_by_age.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("user_data.csv")

# Check and correct column names if needed
print(df.columns)  # Debugging step to check columns

# Show the ages with the highest income
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='age', y='total_income', estimator=sum)
plt.title('Total Income by Age')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.savefig('total_income_by_age.png')
plt.show()

# Show the gender distribution across spending categories
expense_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
for category in expense_categories:
    if category in df.columns:
        df[category] = pd.to_numeric(df[category], errors='coerce')
    else:
        df[category] = 0  # or some default value if the column is missing

df_melted = df.melt(id_vars=["gender"], value_vars=expense_categories, var_name="Expense Category", value_name="Amount")

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='gender', y='Amount' )
plt.title('Spending Distribution by Gender')
plt.xlabel('Expense Category')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.savefig('spending_distribution_by_gender.png')
plt.show()
```
